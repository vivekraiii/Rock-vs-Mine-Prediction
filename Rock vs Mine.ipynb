```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonardata=pd.read_csv('sonar.all-data.csv',header =None)
sonardata.head()
```

```python
sonardata[60].value_counts()
```

```python
sonardata.groupby(60).mean()
```

```python
x=sonardata.drop(columns=60 ,axis=1)
```

```python
x
```

```python
y=sonardata[60]
```

```python
y
```

```python
 sonardata.isnull().sum()
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.1,stratify=y,random_state=1)
```

```python
print(x.shape,xtest.shape,xtrain.shape)
```

```python
xtrain
```

```python
model=LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
x_trp=model.predict(xtrain)
```

```python
accuracy=accuracy_score(x_trp,ytrain)
```

```python
print(accuracy)
```

```python
xtst=model.predict(xtest)
```

```python
accur=accuracy_score(xtst,ytest)
print(accur)
```
